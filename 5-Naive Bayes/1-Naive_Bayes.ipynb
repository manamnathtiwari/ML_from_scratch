Implementing Naive Baye's ML Algorithm Implementation

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X,y = load_iris(return_X_y = True)
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.3,random_state=0)
```

```python
X_train
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

plt.figure(figsize=(10,8))
sns.scatterplot(x=X_train[:,0],y=X_train[:,1],hue=y_train,palette = 'viridis',s=100,alpha=1)
plt.title('Iris Dataset - Training Data Visualization',fontsize=16,fontweight='bold')
plt.xlabel('Sepal Length (cm)',fontsize=12)
plt.ylabel('Sepal Length (cm)',fontsize=12)
plt.legend(title ='Species',labels = ['Setosa','Versicolor','Virginica'])
plt.grid(True,alpha = 0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

### Applying Multinomial Naive Bayes on the Iris Dataset

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train,y_train)
y_predmnb = mnb.predict(X_test)
print(confusion_matrix(y_predmnb,y_test))
print(accuracy_score(y_predmnb,y_test))
print(classification_report(y_predmnb,y_test))
```

#### MultinomialNB is not suitable for continous data like the Iris dataset. It is designed for discrete data, such as word counts in text classification. Using MultinomialNB on continous data can lead to poor performance, as seen in the results above. For continous data , GaussianNB is the appropriate choice 

## Applying Bernouli Naive Bayes 

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(X_train,y_train)
y_predbnb = bnb.predict(X_test)
print(confusion_matrix(y_predbnb,y_test))
print(accuracy_score(y_predbnb,y_test))
print(classification_report(y_predbnb,y_test))
```

##### BernoulliNB is used for binary/boolean features  ,since our data is continous it is not performing well .
