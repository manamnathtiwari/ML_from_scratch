# Decision Tree Classifier


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
# Independent Features 
X = pd.DataFrame(iris['data'],columns=['sepal length','sepal width','petal length','petal wifth'])
```

```python
X.head()
```

```python
# Dependent Features 
y = iris['target']
```

```python
# Train test plot
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 10)
```

```python
# Apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
```

```python
treeClassifier.fit(X_train,y_train)
```

```python
# Visualizing the Decision Tree
from sklearn import tree
plt.figure(figsize =(15,10))
tree.plot_tree(treeClassifier,filled = True)
```

```python
y_pred = treeClassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

## Post Pruning 

### since we constructe decision tree to completely to it's depth it has undergone overfitting , We will now overcome overfitting by post pruning the tree (constructing the whole tree then pruning to certain depth basically to avoid overfitting ). Now i am pruning the above constructed tree to depth 2 because if you carefully observe ,at the depth one category is completely dominating other two categories , then what is the reason to split from there..

```python
post_pruning_tree = DecisionTreeClassifier(max_depth = 2)
post_pruning_tree.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(post_pruning_tree,filled = True)
y_pred = post_pruning_tree.predict(X_test)
```

```python
# Calculating metrics for post pruning tree
accuracy  = accuracy_score(y_test,y_pred)
print(accuracy)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
